```python
# Passive users per week
df_passive = pd.DataFrame(columns=['start_date','total_passive'])
df_passive.start_date = list(weeks.keys())[:13]

for i in np.arange(0,len(weeks)-1,1):
    lower_bound = list(weeks)[i]
    upper_bound = list(weeks)[i+1]
    curr_df = journey_df[(journey_df.index >= lower_bound) & (journey_df.index < upper_bound)]
    
    total_passive = len(curr_df.user_id.unique())-len(curr_df[curr_df.event_type=='confirm_order'].user_id.unique())
    df_passive.total_passive[i] = total_passive
```

# Weekly Activity Recording per User Type

```python
# Create lists of main categories 
frontend_interactions = ['connect_dashboard','check_trades','check_leaderboard','check_feedback']
trading_interactions = ['confirm_order','take_loss','take_loss']
chatbot_interactions = ['read_daily_dose','dose_extra']
```

```python
def activity_per_user_type(d):
    d = d.date()
    dates = [d-timedelta(days=j) for j in range(1,8)]
    lower_bound = min(dates)
    upper_bound = max(dates)
    curr_df = journey_df[(journey_df.index >= lower_bound) & (journey_df.index <= upper_bound)]
    
    
    total_active = curr_df['user_id'].unique()
    total_churn = curr_df[curr_df['event_type']=='churn']['user_id'].unique()
    total_passive = curr_df[~(curr_df.user_id.isin(list(curr_df[(curr_df.event_type=='confirm_order')].user_id.unique())))].user_id.unique()

    final_results = pd.DataFrame(columns=['user_type','perc_trading','perc_chatbot','perc_frontend'])
    final_results.user_type=['active','passive','churn']
    
    count_activity_churn = pd.DataFrame(columns=['user_id','trading','chatbot','frontend'])
    count_activity_churn['user_id'] = total_churn
    count_activity_active = pd.DataFrame(columns=['user_id','trading','chatbot','frontend'])
    count_activity_active['user_id'] = total_active
    count_activity_passive = pd.DataFrame(columns=['user_id','trading','chatbot','frontend'])
    count_activity_passive['user_id'] = total_passive
    
    for i in total_churn:
        user_df = curr_df[curr_df.user_id==i]
        if len(user_df[user_df.event_type.isin(frontend_interactions)])>0:
            count_activity_churn.loc[count_activity_churn.user_id==i, 'frontend']=1
        else:
            count_activity_churn.loc[count_activity_churn.user_id==i, 'frontend']=0
        if len(user_df[user_df.event_type.isin(trading_interactions)])>0:
            count_activity_churn.loc[count_activity_churn.user_id==i, 'trading']=1
        else:
            count_activity_churn.loc[count_activity_churn.user_id==i, 'trading']=0
        if len(user_df[user_df.event_type.isin(chatbot_interactions)])>0:
            count_activity_churn.loc[count_activity_churn.user_id==i, 'chatbot']=1
        else:
            count_activity_churn.loc[count_activity_churn.user_id==i, 'chatbot']=0
    
    s1 = sum(count_activity_churn.trading)+sum(count_activity_churn.chatbot)+sum(count_activity_churn.frontend)
    final_results.loc[final_results.user_type=='churn','perc_trading'] = round(sum(count_activity_churn.trading)/s1*100)
    final_results.loc[final_results.user_type=='churn','perc_chatbot'] = round(sum(count_activity_churn.chatbot)/s1*100)
    final_results.loc[final_results.user_type=='churn','perc_frontend'] = round(sum(count_activity_churn.frontend)/s1*100)
                                                  
    for i in total_active:
        user_df = curr_df[curr_df.user_id==i]
        if len(user_df[user_df.event_type.isin(frontend_interactions)])>0:
            count_activity_active.loc[count_activity_active.user_id==i, 'frontend']=1
        else:
            count_activity_active.loc[count_activity_active.user_id==i, 'frontend']=0
        if len(user_df[user_df.event_type.isin(trading_interactions)])>0:
            count_activity_active.loc[count_activity_active.user_id==i, 'trading']=1
        else:
            count_activity_active.loc[count_activity_active.user_id==i, 'trading']=0
        if len(user_df[user_df.event_type.isin(chatbot_interactions)])>0:
            count_activity_active.loc[count_activity_active.user_id==i, 'chatbot']=1
        else:
            count_activity_active.loc[count_activity_active.user_id==i, 'chatbot']=0

    s2 = sum(count_activity_active.trading)+sum(count_activity_active.chatbot)+sum(count_activity_active.frontend)
    final_results.loc[final_results.user_type=='active','perc_trading'] = round(sum(count_activity_active.trading)/s2*100)
    final_results.loc[final_results.user_type=='active','perc_chatbot'] = round(sum(count_activity_active.chatbot)/s2*100)
    final_results.loc[final_results.user_type=='active','perc_frontend'] = round(sum(count_activity_active.frontend)/s2*100)
                                                  
    for i in total_passive:
        user_df = curr_df[curr_df.user_id==i]
        if len(user_df[user_df.event_type.isin(frontend_interactions)])>0:
            count_activity_passive.loc[count_activity_passive.user_id==i, 'frontend']=1
        else:
            count_activity_passive.loc[count_activity_passive.user_id==i, 'frontend']=0
        if len(user_df[user_df.event_type.isin(trading_interactions)])>0:
            count_activity_passive.loc[count_activity_passive.user_id==i, 'trading']=1
        else:
            count_activity_passive.loc[count_activity_passive.user_id==i, 'trading']=0
        if len(user_df[user_df.event_type.isin(chatbot_interactions)])>0:
            count_activity_passive.loc[count_activity_passive.user_id==i, 'chatbot']=1
        else:
            count_activity_passive.loc[count_activity_passive.user_id==i, 'chatbot']=0

    s3 = sum(count_activity_passive.trading)+sum(count_activity_passive.chatbot)+sum(count_activity_passive.frontend)
    final_results.loc[final_results.user_type=='passive','perc_trading'] = round(sum(count_activity_passive.trading)/s3*100)
    final_results.loc[final_results.user_type=='passive','perc_chatbot'] = round(sum(count_activity_passive.chatbot)/s3*100)
    final_results.loc[final_results.user_type=='passive','perc_frontend'] = round(sum(count_activity_passive.frontend)/s3*100)
    
    
    fig = plt.figure(figsize=(15,15))
    user_types = list(final_results.user_type)
    trading = final_results.perc_trading.values
    chatbot = final_results.perc_chatbot.values
    frontend = final_results.perc_frontend.values

    plt.bar(user_types, frontend, width=0.6, label='% of frontend interactions', color='azure', bottom=chatbot+trading)
    plt.bar(user_types, chatbot, width=0.6, label='% of chatbot interactions', color='skyblue', bottom=trading)
    plt.bar(user_types, trading, width=0.6, label='% of trading interactions', color='darkblue')
    
    plt.xticks(user_types, user_types)
    plt.ylabel("Proportion of user-activity",fontsize=(15))
    plt.xlabel("User Category",fontsize=(15))
    plt.legend(loc="upper right")
    plt.title(f"Activity per user-category for the week before {d}", fontsize=(15))
    plt.show()
    fig.savefig('/Users/alexislacabane/Documents/NewsPill/archive/activity_for_week')
```
